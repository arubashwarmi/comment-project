```python
#First, we import the appropriate modules

import csv

import numpy as np

#Then, we develop a function to find the final grade of a student

def finalgradefinder(studentinfo:list):
    numerator = 0
    denominator = 0
    for grade in studentinfo[14:21]:
        grade = grade.split('/')
        numerator += int(grade[0])
        denominator += int(grade[1])
    finalgrade = (numerator/denominator) * 100
    return finalgrade

#Templates for learning indicator text. I tried to find a way to include that in my spreadsheet but I wasn't able to do it in a way that satisfied me.

def learning_indicator_1_writer(score):
    finalcomment = ""
    base_text_chunk_1 = "You were able to, by the end of the semester, successfully intentionally procrastinate in "
    base_text_chunk_2 = "You should keep working on this in the future. "
    base_text_chunk_3 = "It's not that easy to procrastinate with intention; oftentimes, we are sucked into procrastinating when we intend to do other things. "
    base_text_chunk_4 = "When asked what you were doing and what you intended to do on the daily check ins, "
    base_text_chunk_5 = "By the end of the semester, "
    base_text_chunk_6 = "of your assignments were turned in after 11:58 PM of the due date. "
    if score == '3/5':
        finalcomment = base_text_chunk_1 + "many situations. " + base_text_chunk_2 + base_text_chunk_3 + base_text_chunk_4 + "many " + base_text_chunk_6
    elif score == '4/5':
        finalcomment = base_text_chunk_1 + "most situations. " + base_text_chunk_2 + base_text_chunk_3 + base_text_chunk_4 + "most " + base_text_chunk_6
    elif score == '5/5':
        finalcomment = base_text_chunk_1 + "almost all situations. " + base_text_chunk_2 + base_text_chunk_3 + base_text_chunk_4 + "almost all " + base_text_chunk_6
    return finalcomment

def learning_indicator_2_writer(score:float):
    finalcomment = ""
    base_text_chunk_1 = "In this course, you were tasked with writing essays, both reflections on your own experience with procrastination and commenting on the work of procrastinationologists. Your work in this learning indicator is "
    base_text_chunk_2 = "In essays, I looked for unique insights, evidence of thought, and strong writing. Your insights were strong. "
    base_text_chunk_3 = "Your commentary on the wider effects of procrastination was also strong. "
    base_text_chunk_4 = "My recommendations for improvement would be: "
    base_text_chunk_5 = "Strong work! "
    if score == '3/5':
        finalcomment = base_text_chunk_1 + "strong. " + base_text_chunk_2 + base_text_chunk_3 + base_text_chunk_4 + "keep looking in the texts for the evidence that supports your unique insights. Also, think broadly and make connections across the different areas of procrastinationology and how it's applied in the real world, and really dig into your explanations and analysis. " + base_text_chunk_5
    elif score == '4/5':
        finalcomment = base_text_chunk_1 + "strong, with clearly stated assertions." + base_text_chunk_2 + base_text_chunk_3 + base_text_chunk_4 + "find the best evidence in the text -- it's always there -- to support your points, and flesh out your analyses a little bit more. " + base_text_chunk_5
    elif score == '4.5/5': 
        finalcomment = base_text_chunk_1 + "strong, with clear and very unique assertions." + base_text_chunk_2 + base_text_chunk_3 + base_text_chunk_4 + "find the best evidence in the text -- it's always there -- to support your points, and flesh out your analyses a little bit more. " + base_text_chunk_5
    elif score == '5/5':
        finalcomment = base_text_chunk_1 + "very strong." + base_text_chunk_2 + base_text_chunk_3 + base_text_chunk_4 + "flesh out your analyses a little bit more. " + base_text_chunk_5
    return finalcomment


def learning_indicator_3_writer(score:float):
    finalcomment = ""
    base_text_chunk_1 = "Two key elements of this course were collaboration and thinking together with others. You worked on developing proficiency and aptitude in discussing procrastinationology with everyone and anyone, as well as building off insights from others to formulate an opinion. You also were given opportunities to work with a diverse range of people, not just those who you knew well. "
    base_text_chunk_2 = "Your collaboration skills are "
    base_text_chunk_3 = "You worked on paying attention to others, and you showed "
    base_text_chunk_4 = "In order to improve on this, "
    base_text_chunk_5 = "Collaboration is a valuable skill that will suit you very well in life! "
    if score == '3/5':
        finalcomment = base_text_chunk_1 + base_text_chunk_2 + "strong, with room for improvement. You showed an ability to interact and work very well with a few people; you also participate occasionally in class discussions. " + base_text_chunk_3 + "progress as the semester went on. " + base_text_chunk_4 + "go out of your way to greet and work with those whom you don't know well. Also, jot down the notes from class discussions in your notebook, and try to speak at least once per class in class discussions --- your insights are great. " + base_text_chunk_5
    elif score == '3.5/5':
        finalcomment = base_text_chunk_1 + base_text_chunk_2 + "strong, with room for improvement. You work well with some, and you participate occasionally in class discussions. " + base_text_chunk_3 + "a lot of progress as the semester went on. " + base_text_chunk_4 + "go out of your way to greet and work with those whom you don't know well. Also, jot down the notes from class discussions in your notebook, and try to speak at least once per class in class discussions --- your insights are great. " + base_text_chunk_5
    elif score == '4.5/5': 
        finalcomment = base_text_chunk_1 + base_text_chunk_2 + "strong. You work well with people, and you participate in class discussions. " + base_text_chunk_3 + "a lot of progress as the semester went on. " + base_text_chunk_4 + "continue honing your skills at working with people whom you don't know well. Also, jot down the notes from class discussions in your notebook, and continue to work on speaking up in class when you feel like it --- your insights are great. " + base_text_chunk_5
    elif score == '5/5':
        finalcomment = base_text_chunk_1 + base_text_chunk_2 + "very strong. You work well with almost if not everyone, and are an active participant in class discussions. " + base_text_chunk_3 + "a lot of progress as the semester went on. " + base_text_chunk_4 + "jot down the notes from class discussions in your notebook. " + base_text_chunk_5
    return finalcomment


def learning_indicator_4_writer(score):
    finalcomment = ""
    base_text_chunk_1 = "This learning indicator was focused on annotations of the text and to a lesser extent, notes and thinking about the text. I wanted to see whether or not you were actively engaging with the material, writing your thoughts and questions down, and working with it to gain deeper understanding. "
    base_text_chunk_2 = "Your annotation skills are "
    base_text_chunk_3 = "Your book is "
    base_text_chunk_4 = "To improve on this skill further, you should "
    base_text_chunk_5 = "Taking notes in general is a skill that will serve you well down the road, in meetings, future classes, and many other situations! "
    if score == '3/5':
        finalcomment = base_text_chunk_1 + base_text_chunk_2 + "developing. At the beginning of the semester, you had very basic annotations (and not many of them) but I saw improvement throughout the semester. " + base_text_chunk_3 + "now fairly well annotated. " + base_text_chunk_4 + "work on adding unique discussion insights from the lessons, as well as taking more time to think with the material every night. See me if you have any questions. " + base_text_chunk_5
    elif score == '4/5':
        finalcomment = base_text_chunk_1 + base_text_chunk_2 + "strong, with room for improvement. As the semester went on, I saw improvement in the quality and quantity of your annotations. " + base_text_chunk_3 + "well annotated. " + base_text_chunk_4 + "work on adding unique discussion insights from the lessons, as well as underlining or circling vocabulary that you don't know. See me if you have any questions. " + base_text_chunk_5
    elif score == '5/5':
        finalcomment = base_text_chunk_1 + base_text_chunk_2 + "very strong. As the semester went on, I saw real improvement in the quality and quantity of your annotations. " + base_text_chunk_3 + "very well annotated. " + base_text_chunk_4 + "work on taking more time during the evenings to review in-class annotations and jot down any connections you may see from there. " + base_text_chunk_5
    return finalcomment

#Writing the summary statement using data from the spreadsheet.

    
def summary_and_personal_attribute_writer():
    f = open('Teacher Comments Coding Spreadsheet - Sheet1 (4).csv')
    next(f)
    personalstatementlist = []
    for line in f:
        line = line.split(',')
        name = line[0]
        personalattributes = [line[5:-28]]
        personalstatement = ""
        for list_personal_attributes in personalattributes:
            personalstatement = "You are " + list_personal_attributes[0] + " and " + list_personal_attributes[1] + ". "
            personalstatement = personalstatement + "You " + list_personal_attributes[2] + " and you " + list_personal_attributes[3] + "."
            personalstatement = personalstatement + " Your " + list_personal_attributes[4]
            personalstatementlist.append(personalstatement)
    return personalstatementlist

summary_and_personal_attribute_writer()

#Cleaning up some grammar errors.
    
def cleaner(statementlist:list):
    cleaned_statement_list = []
    for statement in statementlist:
        statement = statement.replace('"', ".")
        statement = statement.replace(".", "")
        statement = statement.strip()
        statement_in_list_form = statement.split()
        cleaned_statement = ""
        for word in statement_in_list_form:
            word = word.strip()
            word = word.replace('.', "")
            cleaned_statement = cleaned_statement + " " + word
        cleaned_statement_list.append(cleaned_statement)
    return cleaned_statement_list
            
cleaner(summary_and_personal_attribute_writer())

#A few more grammar errors... by the end of the time, most of them were cleaned. There were a few which I still couldn't figure out but I left it because I wasn't sure about my own worktime spent on finding the final last bugs. Also, I feel like I tried close to if not everything. Also, some of this code worked when I was testing it out in CoCalc but then bailed on me when I tried to write the file.

def proper_conjugation_and_punctuation(statementlist:list):
    proper_conjugation_list = []
    for comment in statementlist:
        finalcomment = ""
        comment = comment.split()
        for x in range(len(comment)):
            if comment[x] == "Your" and comment[x+1] == "has":
                comment[x] = "You"
                comment[x+1] = "have "   #correcting a common error specifically in this code
            if comment[x][0] == "Y" and x != 0:
                comment[x-1] = comment[x-1] + "." #adding proper punctuation after sentences are finished
            if comment[x][-1] == "s" and comment[x-1] == "you":
                comment[x] = comment[x].replace("s", '')   #correcting incorrect verb conjugations after "You"
            if comment[x] == "Your" and comment[x+1] == "annotations":
                comment[x+1] = "annotations are"   #fixing one specific error
            if comment[x] == "You" and comment[x+1] == "active ":
                comment[x] = "You are"
                comment[x+1] = "an active" #this is to fix another specific error
            finalcomment = finalcomment + " " + comment[x]
        proper_conjugation_list.append(finalcomment)
    print(proper_conjugation_list)
    return proper_conjugation_list


# This commented code here was mainly for me to check to see if the grades and the scores that I gave them all added up. 

#grade_num_dict = {}

    
#with open('Teacher Comments Coding Spreadsheet - Sheet1 (4).csv') as g:
#    f = csv.reader(g)
#    header = next(f)
#    for line in f:
#        grade_num_dict[finalgradefinder(line)] = line[2]
  #  print(grade_num_dict)


#Final, final writer of the comment. This is the ULTIMATE: loads the file, does all the work, and writes to another file.

#I made sure to break up some of these lines of code so it looks cleaner. I could have theoretically done some of them in one giant line.

def comment_writer():
    with open('Teacher Comments Coding Spreadsheet - Sheet1 (4).csv') as g:
        f = csv.reader(g)
        header = next(f)
        x = 0
        for student in f:
            learning_indicator_paragraphs = ""
            summaryparagraph = proper_conjugation_and_punctuation(cleaner(summary_and_personal_attribute_writer()))
            finalgrade = finalgradefinder(student)
            learning_indicator_one_score = student[7]
            learning_indicator_two_score = student[9]
            learning_indicator_three_score = student[11]
            learning_indicator_four_score = student[13]
            learning_indicator_paragraphs = learning_indicator_1_writer(learning_indicator_one_score) 
            learning_indicator_paragraphs = learning_indicator_paragraphs + learning_indicator_2_writer(learning_indicator_two_score)
            learning_indicator_paragraphs = learning_indicator_paragraphs + learning_indicator_3_writer(learning_indicator_three_score)
            learning_indicator_paragraphs = learning_indicator_paragraphs + learning_indicator_4_writer(learning_indicator_four_score)
            finalparagraph = (f"{student[0]}, it was a pleasure to have you in class this semester. Your final grade is {student[2]}. I'm a tough grader though, you know that. All the best for your future, and keep intentionally procrastinating!!")
            completedcomment = summaryparagraph[x] + ". " + learning_indicator_paragraphs + finalparagraph
            completedcomment = completedcomment.strip()
            x += 1
            with open(f'{student[0].title()}_comment.txt', 'w') as h:
                h.write(completedcomment)

comment_writer()
```
